```python
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
background_color = "#FFFFFF"
red = "#ED2939"
purple = "#9400D3"
blue = "#3F00FF"
sns.set(palette=[red, purple, blue], style='white')
```

```python
max_val = 500
game = 'mario'
dqn = np.load('Scores/' + game + '_dqn_scores.npy')[:max_val]
ac = np.load('Scores/' + game + '_ac_scores.npy')[:max_val]
neat = np.load('Scores/' + game + '_neat_scores.npy')[:max_val]

df = pd.DataFrame({'DQN': dqn, 'A2C': ac, 'NEAT': neat})
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 3), dpi=100, facecolor=background_color)
fig.tight_layout
sns.despine()

legend_handles = [patches.Patch(edgecolor="black", facecolor=red, label="DQN"), patches.Patch(edgecolor="black", facecolor=purple, label="A2C"), patches.Patch(edgecolor="black", facecolor=blue, label="NEAT")]

ax_sns = sns.lineplot(data=df, ax=ax, ci=None, legend=False, zorder=2, dashes=False)
ax_sns.set_facecolor(background_color)
ax_sns.set_xlim(0)
ax_sns.tick_params(labelsize=10, width=1, length=2.5)
ax_sns.set_xlabel('Epochs', fontsize=10, weight="bold")
ax_sns.set_ylabel('Score', fontsize=10, weight="bold")

fig.legend(handles=legend_handles, ncol=3, facecolor=background_color, edgecolor=background_color, fontsize=9, loc="upper left", bbox_to_anchor=(0.14, 0.97))
fig.text(0.14, 0.97, "Training Scores on Cartpole", fontsize=18, weight="bold")
```
